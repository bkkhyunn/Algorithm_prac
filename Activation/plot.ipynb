```python
import numpy as np
import matplotlib.pyplot as plt

# Maxout  
def maxout(x, w, b):
    # x:  
    # w:   ( maxout unit  )
    # b: bias
    z = np.dot(x, w) + b
    return z, np.max(z, axis=1)

#   
x = np.linspace(-2, 2, 100).reshape(-1, 1)  # 100 
w = np.random.randn(1, 3)  # 3   
b = np.random.randn(1, 3)  # bias

# Maxout  
z, y_maxout = maxout(x, w, b)

#  
plt.figure(figsize=(8, 6))

#     
for i in range(z.shape[1]):
    plt.plot(x, z[:, i], label=f'Linear {i+1}', linestyle='--')

# Maxout  
plt.plot(x, y_maxout, label='Maxout', color='red', linewidth=2)

#  
plt.title('Maxout Activation Function')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from activation import *
import matplotlib.pyplot as plt
plt.style.use('default')

# Define ELU function and its derivative
def elu_derivative(x, alpha=1.0):
    return np.where(x > 0, 1, alpha * np.exp(x))

x = np.linspace(-5, 5, 400)
alpha_values = [0.1, 0.2, 0.3, 1.0]

plt.figure(figsize=(12, 6))

# Plot ELU and its derivative
for i, alpha in enumerate(alpha_values):
    # ELU function
    y_elu = elu(x, alpha=alpha)
    
    # ELU derivative
    y_elu_prime = elu_derivative(x, alpha=alpha)
    
    # Subplot for ELU function
    plt.subplot(1, 2, 1)
    plt.plot(x, y_elu, label=f'alpha = {alpha}')
    plt.title('ELU Function')
    plt.grid(True)
    plt.legend()
    
    # Subplot for ELU derivative
    plt.subplot(1, 2, 2)
    plt.plot(x, y_elu_prime, label=f'alpha = {alpha}')
    plt.title('ELU Derivative')
    plt.grid(True)
    plt.legend()

plt.show()
```

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

# Swish  
def swish(x, beta=1):
    return x / (1 + np.exp(-beta * x))

beta_values = [0.1, 0.5, 1, 10]
x_vals = np.linspace(-5, 5, 100)

plt.figure(figsize=(12, 6))

for beta in beta_values:
    x = sym.symbols('x')
    swish_expr = x / (1 + sym.exp(-beta * x))  # Swish 
    swish_derivative = sym.diff(swish_expr, x)  #  
    swish_deriv_np = sym.lambdify(x, swish_derivative, 'numpy')  # numpy 
    y = swish(x_vals, beta)  # Swish  
    y_deriv = swish_deriv_np(x_vals)  #  
    
    # Swish  
    plt.subplot(1, 2, 1)
    plt.plot(x_vals, y, label=f'beta = {beta}')
    plt.title('Swish Function')
    plt.grid(True)
    plt.legend()
    
    #  
    plt.subplot(1, 2, 2)
    plt.plot(x_vals, y_deriv, label=f'beta = {beta}')
    plt.title('Swish Derivative')
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
plt.style.use('default')

# Mish  
def mish(x):
    return x * np.tanh(np.log(1 + np.exp(x)))

x_vals = np.linspace(-5, 5, 100)

plt.figure(figsize=(12, 6))

x = sym.symbols('x')
mish_expr = x * sym.tanh(sym.log(1 + sym.exp(x)))
mish_derivative = sym.diff(mish_expr, x)  #  
mish_deriv_np = sym.lambdify(x, mish_derivative, 'numpy')  # numpy 
y = mish(x_vals)  # Mish  
y_deriv = mish_deriv_np(x_vals)  #  

# Mish  
plt.subplot(1, 2, 1)
plt.plot(x_vals, y)
plt.title('Mish Function')
plt.grid(True)

#  
plt.subplot(1, 2, 2)
plt.plot(x_vals, y_deriv)
plt.title('Mish Derivative')
plt.grid(True)

plt.tight_layout()
plt.show()
```
